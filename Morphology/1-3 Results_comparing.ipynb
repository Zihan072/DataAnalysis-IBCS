**One problem: In different method, the sequence number of same cluster maybe are different. For example, in menthod1, for the first cluster, the label is "0". In method 2 , the firt cluster which is similar with the first cluster (label is 0) in menthod 1 maybe will be given the label as "1". In order to compare easily, I will exchange the automatically generated sequence number of clusters in all methods.**

```python
hc_all_list = hc_all_list_not
```

```python
hc_all_list['3 clusters_PCA_ward'].replace(1, 3, inplace=True)
hc_all_list.to_csv("./data/Hc_List_all_after_exchange_position.csv") 
hc_all_list
```

```python
hc_all_list['3 clusters_PCA_ward'].replace(2, 0, inplace=True)
hc_all_list['3 clusters_PCA_ward'].replace(3, 2, inplace=True)
hc_all_list
```

```python
hc_all_list['4 clusters_average'].replace(2, 4, inplace=True)
hc_all_list['4 clusters_average'].replace(1, 2, inplace=True)
hc_all_list['4 clusters_average'].replace(4, 1, inplace=True)
```

```python
hc_all_list.to_csv("./data/Hc_List_all_after_exchange_position.csv") 
```

```python
hc_all_list['6 clusters'].replace(1, 8, inplace=True)
hc_all_list['6 clusters'].replace(0, 1, inplace=True)
hc_all_list['6 clusters'].replace(8, 0, inplace=True)
```

```python
hc_all_list
```

```python
hc_all_list['3 clusters'].replace(2, 5, inplace=True)
hc_all_list['3 clusters'].replace(1, 2, inplace=True)
hc_all_list['3 clusters'].replace(0, 1, inplace=True)
hc_all_list['3 clusters'].replace(5, 0, inplace=True)
hc_all_list
```

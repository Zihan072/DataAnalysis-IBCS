```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
from pandas.plotting import scatter_matrix
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.decomposition import PCA

```

## C1-C88_50µm+SA_no_dead

```python
data = pd.read_csv("./C1-C88_50µm+SA_no_dead.csv")
data.set_index(['Serialnumber (S.No)'], drop=True, inplace=True)
data.index.name = None
data
```

```python
fig = plt.figure(figsize=(15, 20))
heatmap = sns.clustermap(data=data, method='ward', metric='euclidean', 
               row_cluster=True, col_cluster=None, cmap="coolwarm", 
                         vmin=0, vmax=3.5, figsize=(15,25))
heatmap.fig.suptitle("Hierarchy Clustering", fontsize=20)
heatmap.ax_heatmap.set_title("Hierarchical Clustering(ward)", fontsize=20)
plt.savefig("./Morphology/C1-C88_50µm+SA_hierarchical_clustering_ward_3.5.png", dpi=300)
```

```python
fig = plt.figure(figsize=(15, 20))
heatmap = sns.clustermap(data=data, method='ward', metric='euclidean', 
               row_cluster=True, col_cluster=True, cmap="coolwarm", 
                         vmin=0, vmax=0.75, figsize=(15,25))
heatmap.fig.suptitle("Two Dimensions Hierarchy Clustering", fontsize=20)
heatmap.ax_heatmap.set_title("Hierarchical Bi-Clustering(ward)", fontsize=20)
plt.savefig("./Morphology/C1-C88_50µm+SA_hierarchical_clustering_ward_0.75.png", dpi=300)
```

```python
fig = plt.figure(figsize=(15, 20))
heatmap = sns.clustermap(data=data, method='ward', metric='euclidean', 
               row_cluster=True, col_cluster=False, cmap="coolwarm", 
                         vmin=0, vmax=0.75, figsize=(15,25))
heatmap.fig.suptitle("Hierarchy Clustering", fontsize=20)
heatmap.ax_heatmap.set_title("Hierarchical Clustering(ward)", fontsize=20)
plt.savefig("./Morphology/C1-C88_50µm+SA_ward_0.75.png", dpi=300)
```

```python
fig = plt.figure(figsize=(15, 20))
heatmap = sns.clustermap(data=data, method='average', metric='euclidean', 
               row_cluster=True, col_cluster=None, cmap="coolwarm", 
                         vmin=0, vmax=0.75, figsize=(15,25))
heatmap.fig.suptitle("Hierarchy Clustering", fontsize=20)
heatmap.ax_heatmap.set_title("Hierarchical Clustering(average)", fontsize=20)
plt.savefig("./Morphology/C1-C88_50µm+SA_hierarchical_clustering_average_0.75.png", dpi=300)
```

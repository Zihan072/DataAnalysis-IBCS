# quantitative analysis_C1-C88_50µm

```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
from pandas.plotting import scatter_matrix
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.decomposition import PCA
```

```python
data = pd.read_csv("./Morphology_new.csv")
data
```

```python
X = data[['Fish lenght[mm]' , 'eye area [mm²]' ,'yolk sac area [mm²]','yolk sac breadth [mm]']]
plt.rcParams['font.size'] = 11
sns.pairplot(X)
plt.savefig('./Pairwise relationships in the original dataset.png',dpi=300)
#Plot pairwise relationships in a dataset.
```

```python
data_0 = data[data["Fish lenght[mm]"]== 0]
```

```python
data_0.to_csv("./Morphology/Morphology_new_dead.csv", index=None)
```

```python
data_1 = data[data["Fish lenght[mm]"]!= 0]
data_1
```

```python
data_1.to_csv("./Morphology/Morphology_no_dead.csv", index=None)
```

```python
X = data_1[['Fish lenght[mm]' , 'eye area [mm²]' ,'yolk sac area [mm²]','yolk sac breadth [mm]']]
plt.rcParams['font.size'] = 11
sns.pairplot(X)
plt.savefig('./Pairwise relationships(no dead).png',dpi=300)
#Plot pairwise relationships in a dataset.
```

```python
from sklearn import metrics
scores1 = []
for k in range(2,10):
    labels1 = KMeans(n_clusters=k).fit(X).labels_
    score1 = metrics.silhouette_score(X, labels1)
    scores1.append(score1)
    
scores1
```

```python
plt.plot(list(range(2,10)), scores1)
plt.xlabel("Number of Clusters Initialized")
plt.ylabel("Sihouette score")
```

## Normalization

```python
data = pd.read_csv("./Morphology/Morphology_no_dead.csv")
data
```

```python
#eye area Ratio = eye area(mm)/Fish lenght[mm]
#yolk sac areaRatio = yolk sac area [mm²]/Fish lenght[mm]
#yolk sac breadth Ratio = yolk sac breadth [mm]/Fish lenght[mm]
data["eye area Ratio"] = data[["Fish lenght[mm]", "eye area [mm²]"]].apply(lambda x:x["eye area [mm²]"]/x["Fish lenght[mm]"],axis=1)
data["yolk sac area Ratio"] = data[["Fish lenght[mm]", "yolk sac area [mm²]"]].apply(lambda x:x["yolk sac area [mm²]"]/x["Fish lenght[mm]"],axis=1)
data["yolk sac breadth Ratio"] = data[["Fish lenght[mm]", "yolk sac breadth [mm]"]].apply(lambda x:x["yolk sac breadth [mm]"]/x["Fish lenght[mm]"],axis=1)
# data.to_csv("./data/original data with ratio.csv")
data
```

```python
data.to_csv("./Morphology/original data with ratio.csv")
```

```python
ratio_data = data[["Fish lenght[mm]", "eye area Ratio", "yolk sac area Ratio", "yolk sac breadth Ratio"]]
ratio_data
```

```python
plt.rcParams['font.size'] = 8
sns.pairplot(ratio_data)
plt.savefig("./Morphology/pair_relationships_ratio.png", dpi=300)
```

```python
ratio1_data = data[["eye area Ratio", "yolk sac area Ratio", "yolk sac breadth Ratio"]]
ratio1_data
```

```python
plt.rcParams['font.size'] = 8
sns.pairplot(ratio1_data)
plt.savefig("./Morphology/pair_relationships_ratio_nolength.png", dpi=300)
```

## Silhouette Coefficient：

```python
ratio_data
```

```python
from sklearn import metrics
scores = []
for k in range(2,10):
    labels = KMeans(n_clusters=k).fit(ratio_data).labels_
    score = metrics.silhouette_score(ratio_data, labels)
    scores.append(score1)
    
scores
```

## The number of clusters is 2

```python
kmeans2 = KMeans(n_clusters=2).fit(ratio_data)
```

```python
data["clusters2"] = kmeans2.labels_
```

```python
data.sort_values("clusters2").to_csv("./Morphology/ratio_2clusters.csv")
data.sort_values("clusters2")
```

```python
data.groupby("clusters2").mean()
```

```python
#The numbers of samples in each cluster.
clustering_count = pd.DataFrame(data['Serialnumber (S.No)'].groupby(data['clusters2']).count()).T
clustering_ratio = (clustering_count / len(data)).round(2).rename({'counts':'percentage'})
print("The number of samples in each cluster:")
print(clustering_count)
print("#"*30)
print("The percentage of each cluster:")
print(clustering_ratio)
```

```python
plt.rcParams['font.size'] = 11
colors = np.array(['orange', 'green', 'red', 'blue', 'yellow', 'black'])
scatter_matrix(data[["Fish lenght[mm]", "eye area Ratio", "yolk sac area Ratio", "yolk sac breadth Ratio"]],s=100, alpha=1, c=colors[data["clusters2"]], figsize=(15,15))
plt.suptitle("with 2 centroid initialized")
plt.savefig("./Morphology/clustering with 2 clusters.png", dpi=300)
```

## KMeans with PCA

```python
 from sklearn.decomposition import PCA
```

```python
ratio_data
```

```python
pca = PCA(n_components=2)
principalComponents = pca.fit_transform(ratio_data)
principalDf = pd.DataFrame(data = principalComponents, columns = ['principal component 1', 'principal component 2'])
principalDf
```

**Silhouette Coefficient**

```python
scores_pca = []
for k in range(2,10):
    labels_pca = KMeans(n_clusters=k).fit(principalComponents).labels_
    score_pca = metrics.silhouette_score(principalComponents, labels_pca)
    scores_pca.append(score_pca)
    
scores_pca
```

```python
plt.plot(list(range(2,10)), scores_pca)
plt.xlabel("Number of Clusters Initialized with pca")
plt.ylabel("Sihouette score")
```

**After dimensionality reduction, Sihouette score shows 2 and 4 is better.**

## The number of clusters is 2

```python
pca_kmeans2 = KMeans(n_clusters=2).fit(principalDf)
principalDf['pca_clusters2'] = pca_kmeans2.labels_
principalDf
```

```python
pca2_data = pd.concat([data, principalDf], axis=1)
pca2_data = pca2_data.drop(columns = ["clusters2", "clusters6"], axis=1)
pca2_data
```

```python
pca2_data.sort_values("pca_clusters2").to_csv("./Morphology/ratio_pca_2clusters.csv")
pca2_data.sort_values("pca_clusters2")
```

```python
pca2_data.groupby("pca_clusters2").mean().to_csv("./Morphology/mean_ratio_pca_2clusters.csv")
pca2_data.groupby("pca_clusters2").mean()
```

```python
pca_clustering2_count = pd.DataFrame(pca2_data['Serialnumber (S.No)'].groupby(pca2_data['pca_clusters2']).count()).T
pca_clustering2_ratio = (pca_clustering2_count / len(pca2_data)).round(2).rename({'counts':'percentage'})
print("The number of samples in each cluster:")
print(pca_clustering2_count)
print("#"*30)
print("The percentage of each cluster:")
print(pca_clustering2_ratio)
```

```python
x_axis = principalDf['principal component 1']
y_axis = principalDf['principal component 2']
fig = plt.figure(figsize = (6,6))
sns.scatterplot(x_axis, y_axis, hue = principalDf['pca_clusters2'], palette = ['red', 'green'])
plt.title('2 component PCA', fontsize = 20)
plt.savefig("./Morphology/pca_2clusters.png", dpi=300)
plt.show()

```

```python
#用格式化字符串合成
# ‘c{0}*.csv’.format(idx)
# idx 是数字 就是 5-120
# 比如当idx = 30 的时候 ‘c{0}*.csv’.format(idx) =  ‘c30*.csv’
#用的array
x = range(5, 130)
for idx in x:
    n = "C{0}_*.csv".format(idx)
   # print(n)   
    print(r"C:\Users\chen7\Music\IBCS\cleaned_data\{0}".format(n))   
    file_exists = os.path.exists(r"C:\Users\chen7\Music\IBCS\cleaned_data\C{0}_1.csv".format(idx))
    print(file_exists)
    if file_exists:
        file = glob.glob(r"C:\Users\chen7\Music\IBCS\cleaned_data\{0}".format(n))
        print(file)
    #    df = pd.DataFrame()
        df = pd.read_csv(file[0])
        df = df.to_numpy()
        print(df.shape)
        for f in file[1:]:
            dataset = pd.read_csv(f)
            dataset = dataset.to_numpy()
            print(dataset.shape)
            df = np.concatenate((df, dataset), axis=1)
        print(df.shape)
        np.savetxt(r"C:\Users\chen7\Music\IBCS\cleaned_data\new\C{0}.csv".format(idx), df, delimiter = ',')
```

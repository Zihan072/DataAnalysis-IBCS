## DBSCAN

```python
from sklearn.decomposition import PCA
from sklearn.cluster import DBSCAN

pca = PCA(n_components=2)
principalComponents = pca.fit_transform(data)
principalDf = pd.DataFrame(data = principalComponents, columns = ['principal component 1', 'principal component 2'])
```

```python
pca_dbscan = DBSCAN(eps=0.03, min_samples=3).fit(principalDf)
```

```python
labels = pca_dbscan.labels_
principalDf['pca_eps0.03_min3'] = labels
```

```python
fig = plt.figure(figsize = (6,6))
sns.scatterplot(data=principalDf, x='principal component 1', y='principal component 2', hue='pca_eps0.03_min3', palette='Paired')
plt.title('2 component PCA(DBSACN)', fontsize = 20)
plt.savefig("./C5-C129/clustering_png/DBSCAN_pca_eps0.03_min3.png", dpi=300)
plt.show()
```

```python
data_all = pd.read_csv("./C5-C129/data_median_all/data_median_all.csv")
pca_data_all = pd.concat([data_all, principalDf], axis=1)
pca_data_all.drop(['principal component 1', 'principal component 2'], axis=1, inplace=True)
pca_data_all.set_index(['Unnamed: 0'], drop=True, inplace=True)
pca_data_all.index.name = None
pca_data_all
```

```python
pca_data_all.to_csv("./C5-C129/data_clustering/dbscan_eps0.03_min3_no_sort_values.csv")
pca_data_all.sort_values("pca_eps0.03_min3").to_csv("./C5-C129/data_clustering/dbscan_eps0.03_min3.csv")
```

```python
df = pd.read_csv("./C5-C129/data_clustering/dbscan_eps0.03_min3.csv")
df.set_index('Unnamed: 0', drop=True, inplace=True)
df.index.name = None
df.drop('pca_eps0.03_min3', axis=1, inplace=True)
fig = plt.figure(figsize=(15, 28))
heatmap = sns.heatmap(data=df, cmap="coolwarm", vmin=0, vmax=0.12, robust=True)
heatmap.set_title("DBSCAN")
plt.axhline(y=5, linewidth=2, color='black', linestyle='-')
plt.axhline(y=113, linewidth=2, color='black', linestyle='-')
plt.axhline(y=117, linewidth=2, color='black', linestyle='-')
plt.axhline(y=120, linewidth=2, color='black', linestyle='-')
#plt.plot()
plt.savefig("./C5-C129/clustering_png/DBSCAN_4clusters.png", dpi=300)
```

```python
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
import numpy as np
import pandas as pd
import os, glob
import matplotlib.pyplot as plt
import seaborn as sns



data = pd.read_csv("./C5-C129/data_median_all/data_median_all_one_control.csv")
data.set_index(['Unnamed: 0'], drop=True, inplace=True)
data.index.name = None

pca = PCA(n_components=2)
principalComponents = pca.fit_transform(data)
principalDf = pd.DataFrame(data = principalComponents, columns = ['principal component 1', 'principal component 2'])


#%%

pca_kmeans = KMeans(n_clusters=5).fit(principalDf)

#%%

principalDf['pca_clusters'] = pca_kmeans.labels_

#%%

x_axis = principalDf['principal component 1']
y_axis = principalDf['principal component 2']
fig = plt.figure(figsize = (8,8))
sns.scatterplot(x_axis, y_axis, hue = principalDf['pca_clusters'], palette = 'Paired')
plt.title('2 component PCA', fontsize = 20)
#plt.savefig("./C5-C129/data_clustering_without_set1control/kmeans_5clusters", dpi=300)
plt.show()

```

```python
data_all = pd.read_csv("./C5-C129/data_median_all/data_median_all_one_control.csv")
pca_data_all = pd.concat([data_all, principalDf], axis=1)
pca_data_all.drop(['principal component 1', 'principal component 2'], axis=1, inplace=True)
pca_data_all.set_index(['Unnamed: 0'], drop=True, inplace=True)
pca_data_all.index.name = None
#%%
pca_data_all.to_csv("./C5-C129/data_clustering_without_set2control/new/kmeans5_no_sort_values.csv")
pca_data_all.sort_values("pca_clusters").to_csv("./C5-C129/data_clustering_without_set2control/new/kmeans5.csv")



df = pd.read_csv("./C5-C129/data_clustering_without_set2control/new/kmeans5.csv")
df.set_index('Unnamed: 0', drop=True, inplace=True)
df.index.name = None
df.drop('pca_clusters', axis=1, inplace=True)
```

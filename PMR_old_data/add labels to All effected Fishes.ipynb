## add labels to all fishes and generate one file

```python
f_name = df["Unnamed: 0"][0]
#print(f_name)
datasum = pd.read_csv("./C5-C129/data_14features/{0}.csv".format(df["Unnamed: 0"][0]))
#zfill: make C27_0 becomes C27_00 
ilist = [f_name+"_"+str(i).zfill(2) for i in range(len(datasum))]

#build one column in the end.
#datasum['index']=pd.Series(ilist)

#build one index column in the begining, o means the position.
datasum.insert(0, 'index', pd.Series(ilist))
#add the action mode to this file
datasum['action mode']= df["action mode"][0]

for i in df["Unnamed: 0"][1:].index:
    print(i)
    file_name = df["Unnamed: 0"][i]
    print(file_name)
    data = pd.read_csv("./C5-C129/data_14features/{0}.csv".format(file_name))
    ilist = [file_name+"_"+str(i).zfill(2) for i in range(len(data))]
    data.insert(0, 'index', pd.Series(ilist))
    data['action mode'] = df["action mode"][i]
    datasum = pd.concat([datasum, data], axis=0)
    
datasum.to_csv("./C5-C129/all_fished_with_labels/effected_compounds_fishes_labeled.csv", index=None)
```
